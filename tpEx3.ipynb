```python
import numpy as np
from sklearn.preprocessing import StandardScaler
# print("hello")
from numpy.linalg import eigvals
import matplotlib.pyplot as plt
from numpy.linalg import eigh
```

```python

# Load CSV file into a numpy array
data = np.genfromtxt('data/ex3.csv', delimiter=';')

# Print the data
print(data)
```

```python
object= StandardScaler()
scale = object.fit_transform(data) 
print(scale)
```

```python
RX = np.corrcoef(scale, rowvar=False)

print(RX)
```

```python


spectrum = eigvals(RX)

print(spectrum)
```

```python


eigenvalues, eigenvectors = eigh(RX)
```

```python

print("eigenvalues")
print(spectrum)
```

```python

print("eigenvectors")
print(eigenvectors)
```

```python

CX = eigenvectors.T.dot(scale.T).T
print(CX)
```

```python
# Determine the number of principal components to retain
cumulative_variance = np.cumsum(spectrum) / np.sum(eigenvalues)
print("Cumulative variance:", cumulative_variance)
n_components = np.argmax(cumulative_variance >= 0.75) + 1
print("Number of principal components to retain:", n_components)
```
